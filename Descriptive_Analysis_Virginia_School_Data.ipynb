```python
# Code snippet10
# Selecting variables with atleast 80% available data
data80 = data

# Displaying the data80 columns
data80.columns
```

```python
# Code snippet 11
# Setting the predictor variables
predictors = data80[[ 'District',  'City',
       'Zip', 'County', 'Is Title I', 'Is Charter', 'Is Magnet',
       'Is Virtual', 'Number Students', 'Number Full-time Teachers',
       'Student/Teacher Ratio', 'Percent Free/Disc Lunch','Percent African American',
       'Percent American Indian', 'Percent Asian', 'Percent Hispanic',
       'Percent Pacific Islander', 'Percent Two or More Races',
       'Percent White', 'Standard Score']]

# Displaying predictor columns
predictors.columns
```

```python
# Code snippet 13
# Calculating correlations between 'Student/Teacher Ratio' and 'Standard Score'
c1 = data[['Student/Teacher Ratio','Standard Score' ]]

# Displaying correlations
c1.corr()
```

```python
# Code snippet 14
# Displaying Correlations
predictors.corr()
```

```python
# Code snippet 16
# Assigning the correlation matrix
correlation_matrix = predictors.corr()

# Select only the 'Standard Score' row from the correlation matrix
pred_corr = correlation_matrix.loc[['Standard Score']]

# Displaying the correlations for standard score
pred_corr
```

```python
print(""" when we check from the above correlation for standard score we can observe that the top 3 influencing factors on the Standard Scores are:\n 1.Percent Free/Disc Lunch.\n 2.Percent African American.\n 3.Percent White""")
```
